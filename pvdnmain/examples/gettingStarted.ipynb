# Using the PVDN dataset
The PVDN dataset can be used quite easily, despite its sophisticated data structure.
To begin, install the requirements listed in the requirements.txt.

## Importing
PVDNDataset wraps the torch.utils.dataset class, integrating it seamlessly into the pytorch environment.

```python
from pvdn import PVDNDataset
```

Images are loaded using OpenCV 2 by default. To display the images, we'll import matplotlib as well.

```python
import matplotlib.pyplot as plt
```

## Loading the Dataset
With initializing a PVDNDataset instance, the dataset will be loaded (images and keypoints are loaded when accessed to save memory).
We'll print the dataset's statistics to check if everything works as expected:

```python
data = PVDNDataset("PVDN/night/test/")
data.print_summary()
```

## Accessing Images and Keypoints
Data inside the PVDN dataset is structured in sequences. One sequence features at least one approaching vehicle.
The dataset returns the images sequentially, along with metadata and keypoints. The following statement returns
- the image,
- an `info` object, containing metadata and sequence information,
- a list of `vehicle` objects, representing the currently visible vehicles and their instances

Picking a random image yields:

```python
img, info, vehicles = data[100]
print("info: ", info)
print("vehicles: ", vehicles)
plt.imshow(img, cmap="gray")
plt.show()


```
